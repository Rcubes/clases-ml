# 1 GRÁFICO

```python
x = np.linspace(0,4*np.pi,100)   # start,stop,cantidad de muestras
y = np.sin(x)
```

```python
#plt.figure()
plt.plot(x,y) # por defecto el comando plot grafica un linechart
plt.show() # opcional pero si no se usa aparece un texto feo sobre el gráfico.
```

> EL CANVAS BLANCO se conoce como figura, y los Ejes son los Axes. Estos dos son las propiedades principales de un gráfico en matplotlib.

# DIVIDIENDO EL CANVAS

## Declarativo

```python
ax1 = plt.subplot(2,2,1)
plt.plot(y) # por defecto plot genera un linechart
ax2 = plt.subplot(2,2,2)
plt.plot(x,y, c = 'r')
ax3 = plt.subplot(2,2,3)
plt.plot(x,y, c = 'g')
ax4 = plt.subplot(2,2,4)
plt.plot(y, c = 'y')



plt.show()
```

## Orientado a Objetos

```python
fig, ax = plt.subplots(2,2) # tamaño de la grilla
ax[0,0].plot(y)
ax[0,1].plot(x,y, c = 'r')
ax[0,1].set_facecolor('b')
ax[1,0].plot(x,y, c = 'g')
ax[1,1].plot(y, c = 'y')

plt.show()
```

# DIFERENCIAS

![](modifiers.PNG)

```python
ax1 = plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Un título')

ax2 = plt.subplot(2,2,2)
plt.plot(x,y, c = 'r')
plt.xlim(0,4)

ax3 = plt.subplot(2,2,3)
plt.plot(x,y, c = 'g')
plt.xlabel('Este es el eje X')
plt.ylabel('Este es el eje y')

ax4 = plt.subplot(2,2,4)
plt.plot(x,y, c = 'y')
plt.yticks(np.arange(-1, 1.2, step=0.2))
plt.xticks(np.arange(0, 13, step=1))

plt.show()
```

```python
fig, ax = plt.subplots(2,2) # nota para grillas de una columna o 1 fila el índice es sólo uno. Ejemplo ax[0]
#gráficos
ax[0,0].plot(y)
ax[0,1].plot(x,y, c = 'r')
ax[1,0].plot(x,y, c = 'g')
ax[1,1].plot(y, c = 'y')

#modificadores
ax[0,0].set_title('Un título')
ax[0,1].set_xlim(0,4)
ax[1,0].set_xlabel('Este es el eje X')
ax[1,1].set_yticks(np.arange(-1, 1.2, step=0.2))
ax[0,1].set_title('Este es el segundo elemento')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,4, figsize = (10, 5))  # ancho: 10, largo: 5
ax[0,0].plot(y)
ax[0,1].plot(range(100),y, c = 'm') # color magenta
ax[0,2].plot(x,y, c = 'g') # color verde
ax[0,3].plot(y[::10], 'o') # puntos cada 10 observaciones   --- ::10
ax[1,0].plot(y, c = 'r') # color rojo

ax[1,1].plot(x,y, '--') # linea segmentada
ax[1,2].plot(y,'--' ,lw = 5, c = 'y') # ancho de línea 3
ax[1,3].plot(y[::10], '--o') #linea segmentada y puntos
ax[0,0].plot(y+1)
ax[0,0].plot(x, y+1)
ax[1,3].plot(y, c='r')

plt.show()

# el argumento no tiene nombre tiene que ir en orden, 
```

# Scatter Plots / dotplots

```python
from scipy.stats import norm
ruido = norm.rvs(loc = 1.5, scale = 3, size = 100, random_state = 123)
x = np.arange(0,100,1)
y = x + ruido*5


plt.scatter(x,y)
plt.show()
```

```python
ruido
```

```python
fig, ax = plt.subplots(1,4, figsize = (10,3)) 
plt.rcParams['axes.facecolor'] = 'black'   # forma declarativa
ax[0].plot(x,y,'o')
ax[1].scatter(x,y, c = 'y')
ax[2].scatter(x, y, c = ruido, edgecolor = 'k', s = 100)
ax[3].scatter(x,y, c = ruido, s = ruido**2, cmap = 'bwr')
#ax[3].set_rcParams['axes.facecolor'] = 'black'
plt.show()


# mapa de color continua
# discreta semaforo =['rojo', 'verde', 'amarillo'] background = ? como cambio solo un subplot
```

```python
ruido = norm.rvs(loc = 1.5, scale = 20, size = 12000, random_state = 123)
x1 = np.linspace(-10, 10, 1000) 
x2 = np.linspace(-5, 5, 1000) 
x3 = np.linspace(-1,1, 10000)
x = np.hstack([x1,x2,x3])
y = x**2 + ruido

plt.figure(figsize = (5,3))
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize = (5,3))
plt.scatter(x,y, alpha = 0.1, s = 2) # alpha cercano a cero es completamente transparente alpha = 1 es completamente solido, opaco
plt.show()
```

# Histogramas

```python
plt.rcParams['axes.facecolor'] = 'white'
x1 = norm.rvs(size = 100, random_state = 123)
x2 = norm.rvs(size = 1000, random_state = 123)
x3 = norm.rvs(size = 10000, random_state = 123)

fig, ax = plt.subplots(3,1, figsize = (8,10))

ax[0].hist(x1) # x1 tiene que numérico para variables categoricas, no funciona 
ax[1].hist(x2, bins = 1000, color = 'red')
ax[2].hist(x3, bins = 100)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df
```

```python
s = df.Sex.value_counts()
s1 = df.Pclass.value_counts('%')*100
s3 = df.groupby('Embarked').Survived.mean()

fig, ax = plt.subplots(1,3, figsize = (15,5))
ax[0].bar(s.index, s)
ax[0].set_title('Diferencias de Género en el Titanic')
ax[1].bar(s1.index, s1)
ax[1].set_title('Porcentaje de Pasajeros por Clase')
ax[2].barh(s3.index, s3)
ax[2].set_title('Probabilidad de Supervivencia dependiendo del Embarque')
plt.tight_layout()
plt.show()

# rotation del label
```

# Ejes Paralelos / interpretar correctamente los datos

```python
from scipy.stats import randint
high = randint.rvs(100,105, size = 10, random_state = 123)
low = randint.rvs(1,4, size = 10, random_state = 123)

plt.plot(high, c = 'r')
plt.plot(low)
plt.show()
```

```python
plt.plot(high, c = 'r')
plt.show()
```

```python
plt.plot(low)
plt.show()
```

```python
fig = plt.figure() #setea figura
ax1 = plt.gca() # setear un eje 
ax2 = ax1.twinx() # eje gemelo
ax1.plot(high, c = 'r')
ax2.plot(low)
plt.title('Título')
plt.show()
```

```python
fig, ax = plt.subplots(1,4, figsize = (20,5)) # cuando una grilla tiene solo una fila o una columna
ax[0].plot(high, c = 'r')
ax[1].plot(low, '--')
ax[2].plot(high, c = 'r')
ax[2].plot(low)
ax[3].plot(high, c = 'r')
ax2 = ax[3].twinx()
ax2.plot(low, '--')


ax[0].set_title('Primer gráfico')
ax[1].set_title('Segundo gráfico')
ax[2].set_title('Ambos gráficos combinados, misma escala')
ax[3].set_title('Ambos gráficos combinados, distinta escala')
plt.tight_layout() # ajusta el tamaño para que entre en la vista disponible
plt.show()
```
