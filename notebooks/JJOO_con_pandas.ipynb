## Crear 4 nuevos subsets de datos, donde cada uno contenga, respectivamente, a los atletas que ganaron Oro, Plata, Bronce y aquellos que no ganaron medallas.


> NOTA: Función para filtrar por tipo de medalla. También es posible filtrar por aquellos atletas que no obtuvieron medallas.

```python
def subset_medal(medalla):
    
    if medalla is None:  ### if medalla == 'Nulo'
        null_filter = df.Medal.isnull()
        out = df.query('@null_filter')
    else: 
        out = df.query(f'Medal =="{medalla}"')
        
    return out

subset_medal(None)
```

```python
gold = subset_medal('Gold').copy()
silver = subset_medal('Silver').copy()
bronze = subset_medal('Bronze').copy()
na_medal = subset_medal(None).copy()
```

## Generar una nueva columna de nombre 'Female' para cada una de los subsets creados. En esta nueva columna, se debe asignar 1 a las filas correspondientes a una mujer, y 0 a las correspondientes a un hombre.

> Este podría ser un resultado pero está demasiado hardcodeado

```python
gold['Female'] = np.where(gold.Sex == 'F', 1, 0)
silver['Female'] = np.where(silver.Sex == 'F', 1, 0)
gold['Female'] = np.where(gold.Sex == 'F', 1, 0)
```

> Mismo resultado anterior pero escrito de manera más elegante y dinámica

```python
consolidado = [gold, silver, bronze, na_medal]

for dataset in consolidado:
    dataset['Female'] = np.where(dataset.Sex == 'F', 1, 0)
```

## Reportar los 10 primeros países con mayor cantidad de participantes para cada una de los subsets creados.

### Método Hardcoded 

```python
gold.groupby('NOC').Name.nunique().sort_values(ascending = False).head(10)
```

### Utilizando una función

Creando una función que permita resolver el provlema anterior

```python
def top_paises(dataset, n, nombre):
    return dataset.groupby('NOC').Name.nunique().sort_values(ascending = False).head(n).rename(nombre)

top_paises(na_medal, 20, 'Sin Medalla')
```

### Utilizando un `for `

```python
lista_nombres = ['Medallas Oro','Medallas Plata', 'Medallas Bronce', 'Sin Medallas']

for dataset, nombre in zip(consolidado, lista_nombres):
    print(top_paises(dataset, 10, nombre))
```

### Utilizando un List Comprehension 

```python
lista_total = [top_paises(dataset, 10, nombre) for dataset, nombre in zip(consolidado, lista_nombres)]
lista_total[0]
```

### Utilizando un Dictionary COmprehension

```python
dict_total = {nombre: top_paises(dataset, 10, nombre) for dataset, nombre in zip(consolidado, lista_nombres)}
dict_total['Medallas Plata']
```

## Generar una función que permita reportar las promedio entre hombres y mujeres de una columna. 

### Forma Hardcoded

```python
gold.groupby('Sex').Weight.mean()
```

### Creando una función 

```python
def mean_by_sex(dataset, variable):
    return dataset.groupby('Sex')[variable].mean()

mean_by_sex(gold, 'Weight')
```

### Aplicandolo a todas las columnas a la vez 

```python
columnas = df.select_dtypes(np.number).columns
columnas
```

```python
mean_by_sex(gold, columnas)
```

### Aplicandolo a todos los Datasets a la vez usando List Comprehension

```python
lista_finalfinal = [mean_by_sex(dataset, columnas) for dataset in consolidado]
lista_finalfinal[3]
```

### Misma respuesta utilizando Dictionary Comprehension 

```python
{index: mean_by_sex(dataset, columnas) for index, dataset in zip(lista_nombres, consolidado)}
```
